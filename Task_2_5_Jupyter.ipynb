```python
import pandas as pd
```

```python
def custom_path(year):
    return r'names/yob' + str(year) + r'.txt'
```

```python
def get_data_to_df(one_year):
    df_names_year = pd.read_csv(custom_path(one_year), names=['Names', 'Gender', 'Count'])
    df_names_year['Year'] = one_year
    return df_names_year
# get_data_to_df(2012)
```

```python
def all_data_by_years(*args):
    list_years1 = args
    some_df = pd.concat([get_data_to_df(year) for year in list_years1])
    return some_df
res_df = all_data_by_years(2012, 2013, 2015, 2016)
res_df
```

```python
def count_top3(*args):
    data_df = all_data_by_years(*args)
    return data_df.groupby(['Names'])['Count'].agg(['sum']).sort_values(by='sum', ascending=False).head(3)

srt = count_top3(2000, 2001)
srt
```

```python
def count_dynamics(*args):
    data_df = all_data_by_years(*args)
    return data_df.groupby(['Gender', 'Year'])['Count'].agg(['sum']).unstack('Year')

count_dynamics(2005, 2010, 2017)
```
